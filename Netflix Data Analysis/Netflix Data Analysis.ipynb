# Converting Date Time format

```python
df["date_added"] = pd.to_datetime(df["date_added"])
df["day_added"] = df["date_added"].dt.day
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month
df["year_added"].astype(int);
df["day_added"].astype(int);
```

# Data Visualization

## Movies and TV -  Bar Chart

```python
sns.countplot(x="type", data=NetflixDF)
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title("Type")
plt.show()
```

# Movies and TV -  Pie Chart

```python
labels = ["Movies", "TV shows"]
size = NetflixDF["type"].value_counts()
explode = [0,0.1]
colors= plt.cm.Wistia(np.linspace(0,1,2))
plt.rcParams["figure.figsize"] = (9,9)
plt.pie(size,labels=labels,colors=colors, explode=explode, shadow=True, startangle=90)
plt.title("Distribution of Type ", fontsize = 35)
plt.legend()
plt.show()
```

# Ratings of shows and movies

```python
ax = sns.countplot(x="rating", data=NetflixDF)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
fig= plt.gcf()
fig.set_size_inches(13,13)
plt.title = "Rating"
plt.show()
```

# Relation between Type and Rating

```python
plt.figure(figsize=(13,10))
sns.countplot(x='rating', hue="type", data=NetflixDF)
plt.title = "Relation between Type and Rating"
plt.show()
```

# Pie-chart for Rating

```python
NetflixDF['rating'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(10,10))
```

# For Country 

```python
plt.subplots(figsize=(20,10))
wordcloud = WordCloud( background_color="black", 
                      width=1920,
                      height = 1080,).generate(" ".join(df.country))

plt.imshow(wordcloud) 
plt.axis("off")
plt.savefig('country_wordcloud.png')
plt.show()          
```
